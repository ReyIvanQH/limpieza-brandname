```python
# ============================================================
# GP Homologation Application v2.0
# Section 1 - Configuration
# ============================================================

from pathlib import Path

# -----------------------------
# Input files
# -----------------------------

INPUT_CUSTOMER_CATALOGUE = Path(r"C:\Users\Ivan.DeLaFuente\OneDrive - Quaker Houghton\Documents\QH WorkDay\July 2026\Industrial Market\Industrial_Market.csv")

INPUT_EXACT_GROUPS = Path("02_exact_normalized_groups.csv")

INPUT_REVIEWED_MATCHES = Path("04_fuzzy_matches_reviewed.csv")


# -----------------------------
# Output files
# -----------------------------

OUTPUT_GP_MAPPING = Path("05_gp_mapping.csv")

OUTPUT_CUSTOMER_CATALOGUE = Path("Customer Catalogue Clean.csv")

OUTPUT_REPORT = Path("Cleaning Industrial Report.txt")


# -----------------------------
# Review decisions
# -----------------------------

DECISION_NO = "NO"
```

```python
# ============================================================
# Section 2 - Imports
# ============================================================

from collections import defaultdict
import pandas as pd
```

```python
# ============================================================
# Section 3 - Load input files
# ============================================================

print("Loading input files...")

df_catalogue = pd.read_csv(INPUT_CUSTOMER_CATALOGUE)

df_exact = pd.read_csv(INPUT_EXACT_GROUPS)

df_review = pd.read_csv(INPUT_REVIEWED_MATCHES)

print("Customer catalogue:", len(df_catalogue), "rows")
print("Exact groups:", len(df_exact), "groups")
print("Reviewed matches:", len(df_review), "candidate pairs")
```

```python
# ============================================================
# Section 4 - Validate input files
# ============================================================

CATALOGUE_COLUMNS = {
    "Global Parent"
}

EXACT_COLUMNS = {
    "Representative",
    "Members"
}

REVIEW_COLUMNS = {
    "Representative_1",
    "Representative_2",
    "Decision"
}




def validate_columns(df, required_columns, dataset_name):

    missing = required_columns - set(df.columns)

    if missing:

        raise ValueError(
            f"{dataset_name} is missing required columns: "
            f"{sorted(missing)}"
        )
    



validate_columns(
    df_catalogue,
    CATALOGUE_COLUMNS,
    "Customer Catalogue"
)

validate_columns(
    df_exact,
    EXACT_COLUMNS,
    "Exact Groups"
)

validate_columns(
    df_review,
    REVIEW_COLUMNS,
    "Reviewed Matches"
)

print("All required columns found.")
```

```python
# ============================================================
# Section 5 - Prepare reviewed decisions
# ============================================================

df_review = df_review.copy()

df_review["Decision"] = (
    df_review["Decision"]
    .fillna("")
    .astype(str)
    .str.strip()
)

approved_merges = df_review[
    (df_review["Decision"] != "")
    &
    (df_review["Decision"].str.upper() != DECISION_NO)
].copy()

print("Approved homologations:", len(approved_merges))
```
