## Limpieza de CN/GP cuando tienen el mismo nombre pero diferente ubicación

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Ivan.DeLaFuente\OneDrive - Quaker Houghton\Documents\QH WorkDay\Junio 2026\T&P Market\T&P_Market.csv')
```

```python
# Create unique location
df['Ship To Country'] = df['Ship To Country'].fillna('')
df['Ship To State']   = df['Ship To State'].fillna('')
df['Ship To City']    = df['Ship To City'].fillna('')

df['Location'] = (
    df['Ship To Country'] + '|' +
    df['Ship To State'] + '|' +
    df['Ship To City']
)

df["Location"] = df["Location"].str.upper()
```

```python
# Group and count unique locations
df["Customer Name Norm"] = df["Customer Name"].str.upper().str.strip()

grouped = df.groupby('Customer Name Norm')['Location'].nunique()
```

```python
# Filter problematic customers
companies_problem = grouped[grouped > 1]
```

```python
# Create the mask
mask = df['Customer Name Norm'].isin(companies_problem.index)
```

```python
# Apply the mask
df_problemas = df[["Global Parent", "Customer Name", "Ship To Country", "Ship To State", "Ship To City"]][mask]
```

```python
df.loc[mask, 'Customer Name Clean'] = (
    df.loc[mask, 'Customer Name'] + ' ' + df.loc[mask, 'Ship To City']
)

df["Customer Name Clean"] = df["Customer Name Clean"].fillna(df["Customer Name"])

changes = df[df["Customer Name"] != df["Customer Name Clean"]][["Customer Name", "Customer Name Clean"]]
```
