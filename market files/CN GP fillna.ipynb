Este codigo agrupa los Global Parent y rellena el valor en diferentes columnas si es que solo hay un valor unico.

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Ivan.DeLaFuente\OneDrive - Quaker Houghton\Documents\QH WorkDay\July 2026\Industrial Market\Industrial_Market_Final.csv')
df_original = df.copy()
```

```python
df['GP Country'] = df['GP Country'].str.title()
df['GP State'] = df['GP State'].str.title()
df['GP City'] = df['GP City'].str.title()
df['GP Super Region'] = df['GP Super Region'].str.title()
df['GP Region'] = df['GP Region'].str.title()
df['GP Sub Region'] = df['GP Sub Region'].str.title()

df['CN Continent'] = df['CN Continent'].str.title()
df['Ship To Country'] = df['Ship To Country'].str.title()
df['Ship To State'] = df['Ship To State'].str.title()
df['Ship To City'] = df['Ship To City'].str.title()
df['CN Super Region'] = df['CN Super Region'].str.title()
df['CN Region'] = df['CN Region'].str.title()
df['CN Sub Region'] = df['CN Sub Region'].str.title()
```

```python
def fill_if_unique(series):
    non_null = series.dropna()
    
    # Solo rellenar si hay exactamente 1 valor único
    if non_null.nunique() == 1:
        return series.fillna(non_null.iloc[0])
    
    # Si hay conflicto o todo vacío → no tocar
    return series
```

### Global Parent

```python
cols = [
    'GP Country',
    'GP State',
    'GP City',
    'GP Latitude',
    'GP Longitude',
    'GP Super Region',
    'GP Region',
    'GP Sub Region'
]

for col in cols:
    df[col] = df.groupby('Global Parent')[col].transform(fill_if_unique)
```

```python
changes = []

for col in cols:
    temp = df[df[col] != df_original[col]][['Global Parent']].copy()
    temp['Column'] = col
    temp['Old Value'] = df_original.loc[temp.index, col]
    temp['New Value'] = df.loc[temp.index, col]
    
    changes.append(temp)

changes_df = pd.concat(changes)
```
