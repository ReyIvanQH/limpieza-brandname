```python
import pandas as pd

market = pd.read_csv(r'C:\Users\Ivan.DeLaFuente\OneDrive - Quaker Houghton\Documents\QH WorkDay\Junio 2026\T&P Market\T&P_Market.csv')
```

```python
# Catalogue Format
catalogue = market.drop(columns=['llave', 'Ship To Number', 'Channel'])
catalogue = catalogue.drop_duplicates(subset=['Customer Name'])
```

```python
# Catalogue IDs
ID_list = []
for i in range (1, len(catalogue)+1):
    new_id = 'T&P' + str(i).zfill(4)
    ID_list.append(new_id)

catalogue['ID'] = ID_list  
```

```python
# Market IDs
market = market.merge(
    catalogue[['Customer Name', 'ID']],
    how='left',
    on='Customer Name'
)

market['ID_x'] = market['ID_y']
market = market.drop(columns=['ID_y'])
market = market.rename(columns={'ID_x': 'ID'})
```

```python
market.to_csv('T&P_Market_Final.csv', index=False)
catalogue.to_csv('T&P_Catalogue.csv', index=False)
```
